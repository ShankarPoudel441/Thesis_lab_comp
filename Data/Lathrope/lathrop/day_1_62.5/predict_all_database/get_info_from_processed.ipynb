```python
DATA_DIR = Path('/home/shankar/GRA/R@/rica/projects/lathrop/day_1_62.5/predict_all_database/oct_14_interpolate_model')
```

```python
cores_df = pd.read_csv(DATA_DIR / 'lathrop-waypoints.csv')
cores_df.head()
```

```python
vib_df = pd.read_csv(DATA_DIR / 'processed_data.csv')
vib_df.head()
```

```python
for index, core in cores_df.iterrows():
    name = core.Name
    print(f"{name}\n=============")
    snapshot = vib_df[(abs(vib_df.latitude - core.latitude) < 0.00001) & (abs(vib_df.longitude - core.longitude) < 0.000013)]
    for i in snapshot.passID.unique():
        # start= snapshot[snapshot.passID == i].time.min()
        # stop=snapshot[snapshot.passID == i].time.max()
        avg=snapshot[snapshot.passID == i].density.mean()
        myMin=snapshot[snapshot.passID == i].density.min()
        myMax=snapshot[snapshot.passID == i].density.max()
        std=snapshot[snapshot.passID == i].density.std()
        # print(f"\tPass {i}: start-> {start} stop-> {stop} density->{estimate:0.1f}")
        print(f"\tPass {i}: min-> {myMin:0.1f} | mean-> {avg:0.1f} +/- {std:0.1f} | max->{myMax:0.1f}")
        # plt.figure()
        # snapshot[snapshot.passID == i].plot(x='time', y='density')
        # plt.savefig(f"{name}_pass{i}.png")
        # plt.close()
```
